````python
import pandas as pd
import numpy as np
````

### 데이터프레임 고급 인덱싱 
판다스 데이터프레임 인덱싱은 라벨, 라벨 리스트, 행 인덱스 슬라이스의 인덱싱 형태 뿐만 아니라 numpy와 같은 `,`를 이용한 `(행 인덱수, 열 인덱스)` 형식의 2차원 인덱스도 제공함 

- `loc` 속성 : 라벨 기반의 2차원 인덱서
- `iloc` 속성 : 순서를 나타내는 정수 기반의 2차원 인덱서

### `loc` 인덱서
행 인덱스 값과 열 인덱스 값을 이용하여 인덱싱을 해주는 인덱서 

```python
df.loc[행인덱스값]
df.loc[행인덱스값, 열인덱스값]
```

````python
df = pd.DataFrame(np.arange(10, 22).reshape(3, 4), index=['a', 'b', 'c'], columns=['A', 'B', 'C', 'D'])
df
````

````python
df.loc['a']
````

````python
df.loc[['a', 'b']]
````

````python
df.loc['a' : 'b']
````

````python
df.loc[['a']]
````

````python
df.loc[df.A > 13]
````

인덱스 값으로 열 인덱스 값을 반환하는 함수를 사용할 수 있음 

````python
def select_rows(df):
    return df.A > 12
````

````python
df.loc[select_rows(df)]
````

`loc` 없이 인덱싱을 할때는 `[]`에 열 인덱스에 대한 인덱싱을 하였는데 `loc`는 열 인덱싱으로 지정할 수 없음 

````python
df.loc['A']
````

`loc`인덱싱에서 정수로 되어있는 행이라고 할 지라도 슬라이싱시에 종료 인덱스 값에 대한 결과도 포함하여 출력됨 

````python
df2 = pd.DataFrame(np.arange(10, 26).reshape(4, 4), columns=['A', 'B', 'C', 'D'])
df2
````

````python
df2.loc[0:2]
````

#### 인덱싱 값으로 행과 열 모두 지정
`loc[행인덱스, 열인덱스]`로 행과 열에 대한 인덱싱을 할 수 있음 

````python
df.loc['a','A']
````

````python
df.loc['b':, :'C']
````

````python
df.loc[df.A > 13, ['B', 'C']]
````

### `iloc` 인덱서
`iloc` 인덱서는 `loc` 인덱서와 사용방법은 동일하나 인덱스로 지정하는 값이 순서를 나타내는 정수 값임 

````python
df
````

````python
df.iloc[0, 1] # 0번째 행에서 1번값 출력
````

````python
df.iloc[:2, 1:]
````

````python
df.iloc[2]
````

##### 파이썬으로 다음 연산을 수행한다.

1. real_estate 데이터베이스의 officetel_estate 테이블을 데이터프레임으로 가져옴
2. 상위 10개에 대한 레코드를 가지는 데이터프레임을 생성 
3. locale이 seoul-downtown인 레코드를 가지는 데이터프레임을 생성 
4. between_lease_and_lease_deposit_ratio가 10이상인 레코드의 locale, base_month 컬럼만 가지는 데이터프레임을 생성 

````python
import mysql.connector
````

````python
# real_estate 데이터베이스의 officetel_estate 테이블을 데이터프레임으로 가져옴

practice_df = None

try:
    conn = mysql.connector.connect(
        host="127.0.0.1",        
        user="root",   
        password="root",
        database="real_estate" 
    )

    if conn.is_connected():
        SQL = 'SELECT * FROM officetel_estate'
        practice_df = pd.read_sql_query(SQL, conn, index_col='sequence')

except Exception as e:
    pass
finally:
    if conn.is_connected():
        conn.close()

practice_df
````

````python
# 상위 10개에 대한 레코드를 가지는 데이터프레임을 생성 
practice_df.iloc[:10]
````

````python
# locale이 seoul-downtown인 레코드를 가지는 데이터프레임을 생성 
practice_df.loc[practice_df.locale=='seoul-downtown']
````

````python
# between_lease_and_lease_deposit_ratio가 10이상인 레코드의 locale, base_month 컬럼만 가지는 데이터프레임을 생성 
practice_df.loc[practice_df.between_lease_and_lease_deposit_ratio >= 10, ['locale', 'base_month']]
````
